The issue for marketing teams arises here; they may know who they are attempting to reach, but they want to know what handful of advertising they should create to appeal to a broad range of individuals without splintering their campaign into thousands of tailored ads, which may be prohibitively expensive.

### creating a function to generate basic information of the data

```python
def data_info(data):
    print("Dataset has",data.shape[1], "colummns and ",data.shape[0],"rows")
    print("Data has ",data[data.duplicated()].shape[0], " duplicate rows")
    print("Dataset missing values:\n",data.isnull().sum()[data.isnull().sum()>0] )
    print ("Dataset attributes datatypes:\n", data.dtypes)
    
```

### Data Cleaning

```python
market_data_copy= market_data
```

```python
market_data_copy.head()
```

#### Renaming the Income column as it has spaces

```python
market_data_copy.rename(columns={' Income ':'Income'},inplace=True)
```

```python
market_data_copy.columns
```

#### Removing $,.,00 signs form Income columns data

```python
market_data['Income']= market_data['Income'].str.strip('$')
market_data['Income']= market_data['Income'].str.replace('.','')
market_data['Income']= market_data['Income'].str.replace(',','')
market_data['Income']= market_data['Income'].str.replace('00','')
```

```python
market_data['Income'][:5]
```

### Converting Income column to integer column

#### Divinding the dataframe to 2 with one have all Income values and other with missing income values

```python
market_data_Income= market_data[market_data.Income.isnull()==False]
```

```python
market_data_Income.isnull().sum()### checking if the new  dataframe doesnot have any missing values
```
